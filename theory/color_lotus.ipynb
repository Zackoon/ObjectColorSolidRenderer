```python
import numpy as np
import matplotlib.pyplot as plt
from chromalab.observer import Observer
from chromalab.spectra import Spectra
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget

# Get cone response functions
standard_trichromat = Observer.trichromat(np.arange(390, 701, 3))
l_response = standard_trichromat.sensors[2].data
m_response = standard_trichromat.sensors[1].data
s_response = standard_trichromat.sensors[0].data

wave_lengths = np.arange(390, 701, 3)

# Create the color locus
locus = np.array(list(zip(l_response, m_response, s_response)))

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the color locus
ax.plot(locus[:, 0], locus[:, 1], locus[:, 2], 'b-')

# Add scatter points colored by wavelength
scatter = ax.scatter(locus[:, 0], locus[:, 1], locus[:, 2], c=wave_lengths, cmap='Spectral')

# Set labels and title
ax.set_xlabel('L Response')
ax.set_ylabel('M Response')
ax.set_zlabel('S Response')
ax.set_title('Color Locus in LMS Space')

# Add a color bar
plt.colorbar(scatter, label='Wavelength (nm)')

# Show the plot
plt.show()
```
